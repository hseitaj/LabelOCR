```python
from mmocr.apis import MMOCRInferencer
import matplotlib.pyplot as plt
import mmcv
import numpy as np

# Provide the path to your image file
imagePath = "C:/Users/hseit/Desktop/SummerMCREU/images/IMG-4233.jpg"

# Create an MMOCRInferencer with 'dbnetpp' for detection and 'svtr-small' for recognition
infer = MMOCRInferencer(det='dbnetpp', rec='svtr-small')

# Use the inferencer on your image
result = infer(imagePath, save_vis=True, return_vis=True)

# Print predictions
print(result['predictions'][0]['rec_texts'])

# Access the visualization image data
visualization_data = result['visualization'][0]

# Convert the visualization data to a NumPy array
image_array = np.array(visualization_data)

# Set the figure size for larger display
plt.figure(figsize=(17, 12))

# Visualize the image with adjusted aspect ratio
plt.imshow(image_array, aspect='auto')
plt.axis('off')
plt.show()

# Visualize the original image
predicted_img = mmcv.imread(imagePath)
plt.imshow(mmcv.bgr2rgb(predicted_img))
plt.axis('off')
plt.show()
```

```python
import cv2
from mmocr.apis import MMOCRInferencer
import matplotlib.pyplot as plt
import numpy as np

# Provide the path to your image file
imagePath = "C:/Users/hseit/Desktop/SummerMCREU/images/IMG-4235.jpg"

# Read the image using OpenCV
image = cv2.imread(imagePath)

# Resize the image using OpenCV
resized_image = cv2.resize(image, (900, 900))

# Convert the image to RGB format
rgb_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)

# Create an MMOCRInferencer with 'dbnetpp' for detection and 'svtr-small' for recognition
infer = MMOCRInferencer(det='dbnetpp', rec='svtr-small')

# Use the inferencer on your resized image
result = infer(rgb_image, save_vis=True, return_vis=True)

# Print predictions
print(result['predictions'][0]['rec_texts'])

# Visualize the image
plt.imshow(rgb_image)
plt.show()

# Visualize the original image
predicted_img = mmcv.imread(imagePath)
plt.imshow(mmcv.bgr2rgb(predicted_img))
plt.show()
```

```python
from mmocr.apis import MMOCRInferencer
import matplotlib.pyplot as plt
import mmcv
imagePath = "C:/Users/hseit/Desktop/SummerMCREU/images/IMG-4286.jpg"

infer = MMOCRInferencer(rec='svtr-small')
result = infer(imagePath, save_vis=True, return_vis=True)

print(result['predictions'])

# Visualize the return value
plt.imshow(result['visualization'][0])
plt.show()

# Visualize the saved image
predicted_img = mmcv.imread(imagePath)#'results/vis/demo_text_recog.jpg')
plt.imshow(mmcv.bgr2rgb(predicted_img))
plt.show()
```

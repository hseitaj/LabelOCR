```python
import os
import re
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import pytesseract
from PIL import Image
import string
from nltk.corpus import stopwords
```

```python
# Set up Tesseract OCR
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
# Path to the folder containing JPG images
folder_path = "C:/Users/hseit/Desktop/SummerMCREU/images/"
amount = 0

# Iterate over the PNG images in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".jpg"):
        amount += 1
        print("File found: " + str(amount))
        # Load the label image
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path)
        print(image_path)
        
        # Preprocess the image
        resized_image = cv2.resize(image, (800, 600))  # Resize the image
        gray = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
        blurred_image = cv2.GaussianBlur(gray, (5, 5), 0)  # Apply Gaussian blur
        _, thresholded_image = cv2.threshold(blurred_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)  # Apply thresholding

        # Perform text detection
        contours, _ = cv2.findContours(thresholded_image.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        text_regions = []
        for contour in contours:
            x, y, w, h = cv2.boundingRect(contour)
            if w > 10 and h > 10:  # Filter out small contours
                text_regions.append((x, y, w, h))

        # Perform text recognition on each region
        extracted_text = []
        for region in text_regions:
            x, y, w, h = region
            text_region = resized_image[y:y+h, x:x+w]
            text = pytesseract.image_to_string(text_region)
            extracted_text.append(text)

        # Display the results
        for region in text_regions:
            x, y, w, h = region
            cv2.rectangle(resized_image, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.imshow("Text Detection", resized_image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

        # Print extracted text
        print("Extracted Text:")
        for text in extracted_text:
            print(text)
```
